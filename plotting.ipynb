### Mean number of driver per cell (n)

```python
df_output_orig = pd.read_csv('InvasiveK512/output.dat', '\t')
df_output_rand = pd.read_csv('InvasiveK512_Asymetric/InvasiveK512_Asymetric_Random/output.dat','\t')
df_output_remove1 = pd.read_csv('InvasiveK512_Asymetric/InvasiveK512_Asymetric_Remove1Neighbor/output.dat', '\t')
df_output_remove2 = pd.read_csv('InvasiveK512_Asymetric/InvasiveK512_Asymetric_Remove2Neighbor/output.dat', '\t')
df_output_remove3 = pd.read_csv('InvasiveK512_Asymetric/InvasiveK512_Asymetric_Remove3Neighbor/output.dat', '\t')
```

```python
df_output_list = [df_output_orig, df_output_rand, df_output_remove1, df_output_remove2, df_output_remove3]
label_list = ["Invasive", 
              "Invasive - Random Removal (0~3)", 
              "Invasive - Remove one neighbor", 
              "Invasive - Remove two neighbor", 
              "Invasive - Remove three neighbor"
             ]
```

```python
for i in range(len(df_output_list)):
    df = df_output_list[i]
    df['Generation'] = df['Generation'].apply(np.floor).astype(int)

    label = label_list[i]
    
    plt.plot(df['Generation'], df['Drivers'], label=label)

plt.xlabel('Generation')
plt.ylabel('Mean number of driver mutation per cell, n')
plt.legend()
```

### Combine Diversity (D) and mean driver per cell (n)

```python
# Define lower bound of D from clonal sweep
def D_clonal_sweeps(n):
    """Trajectory for clonal sweeps."""
    fractional_part = n % 1
    return 1 / ((1 - fractional_part)**2 + fractional_part**2)

def D_branching_curve(n):
    """Intermediate branching curve approximation."""
    return 9 * (2 * n - 1) / 8
```

```python
for i in range(len(df_list)):

    df_inv = df_list[i]
    df_output = df_output_list[i]
    
    df_inv['Generation'] = df_inv['Generation'].apply(np.floor).astype(int)
    df_output['Generation'] = df_output['Generation'].apply(np.floor).astype(int)
    
    df_combined = pd.merge(
        left=df_inv[['Generation', 'DriverDiversity']],
        right=df_output[['Generation', 'Drivers']],
        on='Generation'
    )
    
    label = label_list[i]
    
    plt.plot(df_combined['Drivers'], df_combined['DriverDiversity'], label=label)

# Plot lower bound and approximation
n_values = np.linspace(0, 4, 500)  # n >= 1
D_clonal = D_clonal_sweeps(n_values)
D_branching = D_branching_curve(n_values)

plt.plot(n_values, 
         D_clonal, 
         #label="Sequential Clonal Sweep", 
         color="gray")
plt.plot(n_values, D_branching, 
         #label="Intermediate Branching Curve", 
         color="gray", linestyle="--")

plt.ylim(1, 12)
plt.xlabel('Mean number of driver mutation per cell, n')
plt.ylabel('Clonal Diversity, D')
plt.legend()
```
